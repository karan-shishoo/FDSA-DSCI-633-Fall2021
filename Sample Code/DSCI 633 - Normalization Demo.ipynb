### Preprocess data with Min-Max Scaling

- X_new = (X - X_min)/(X_max - X_min)
- Sometimes called "Scale Normalization", scales the range to [0, 1]
- Squishes the n-dimensional data into an n-dimensional unit hypercube
- Highly sensitive to outliers


```python
scaler = MinMaxScaler()
scaled_data = data
scaled_data[['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']] = scaler.fit_transform(data[['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']])  
print(scaled_data.head())
plt.plot(scaled_data['Sepal Length'])
plt.show()
sb.kdeplot(scaled_data['Sepal Length'])
```

### Preprocess data with Standard Scaling

- X_new = (X - mean)/Std
- Also called "Z-Score Normalization", translates mean to 0 and scales such that std dev is 1
- Less sensitive to outliers



```python
scaler = StandardScaler()
scaled_data[['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']] = scaler.fit_transform(data[['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']])  
print(scaled_data.head())
plt.plot(scaled_data['Sepal Length'])
plt.show()
sb.kdeplot(scaled_data['Sepal Length'])
```
