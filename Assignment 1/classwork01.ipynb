```python
import pandas as pd
```

```python
df = pd.read_csv('iris.csv', header = None)
```

```python
df.columns = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width', 'Class']
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
uniques = list(df['Class'].unique())
df['Class_number'] = [uniques.index(i) for i in df['Class']]
```

```python
plt.plot()
plt.scatter(df["Sepal_length"], df["Sepal_width"], c=df['Class_number'])
```

```python
import seaborn as sns
```

```python
sns.pairplot(df, hue='Class')
```
